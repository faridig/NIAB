### nettoyage de bande annonce YT

```python
df_yt = pd.read_csv('YT_API/video_details.csv')
```

```python
#supprimer dans le titre Bande annonce et ce qu'il y a après
import re 

df_yt['Title'] = df_yt['Title'].apply(lambda x: re.sub(r'(?i)\sBande\sAnnonce.*', '', x))

```

```python
# supprimer tout ce qui vient après les : 

df_yt['Title'] = df_yt['Title'].apply(lambda x: re.sub(r':.*', '', x))

df_yt.head()
```

```python
# changement date 

df_yt['Published_date'] = df_yt['Published_date'].str.split('-').str[0]

df_yt.rename(columns={'Published_date': 'Release'}, inplace=True)
```

```python

df_yt = df_yt[df_yt['Tags'].str.contains('cinema', case=False, na=True)]

df_yt.head(50)

```
